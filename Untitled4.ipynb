# Kody pomocnicze do zadań z  lab s(zero)

```python
def SZero(a):
  result = "zero"
  for i in range(0,a):
    result = "s("+result+")" 
  return result
def SZeroToNumber(a):
  result = 0
  while a != 'zero':
    result+=1
    a = a[2:-1]
  return result
```

```python
print(SZero(8))
```

```python
print(SZero(5))
```

```python
SZeroToNumber('s(s(s(s(s(zero)))))')
```

## Kod generujący bicia konika

```python
def knight_take():
  bicia = [[-2,-1],[-1,-2],[-2,1],[-1,2],[2,-1],[1,-2],[2,1],[1,2]]
  for b in range(0,len(bicia)):
    for i in range(0,8):
      for j in range(0,8):
        result = "knight_take(A,B,c("
        for k in range(0,8):
          for l in range(0,8):
            if k != 0 and l == 0:
              result = result +'                  ' #18
            if k == i and l == j:
              result = result + 'A,'
            elif k == i + bicia[b][0] and l==j + bicia[b][1]:#waunek
              result = result + 'B,'
            else:
              result = result + '_,'
            if l == 7 and k != 7:
              result = result + '\n'
        result = result[:-1]+')).'
        if result.count('B')==2:
          print(result)

```

### Wynik

```python
knight_take()
```

## Kod generujący krawędzie

```python
def edge():
  for i in range(0,8):
    for j in range(0,8):
      result = "edge(A,c("
      for k in range(0,8):
        for l in range(0,8):
          if k != 0 and l == 0:
            result = result +'         ' #9
          if k == i and l == j and (k==0 or k == 7 or l == 0 or l == 7):
            result = result + 'A,'
          else:
            result = result + '_,'
          if l == 7 and k != 7:
            result = result + '\n'
      result = result[:-1]+')).'
      if result.count('A')==2:
        print(result)
```

### Wynik

```python
edge()
```
